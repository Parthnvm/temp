```python
import matplotlib.pyplot as plt
```

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
kohli = [950, 1100, 1200, 1350, 1250, 980, 1020, 1150, 1080, 1180]
rohit = [780, 850, 970, 1050, 1150, 900, 990, 1040, 1120, 1200]
sehwag = [620, 650, 700, 720, 680, 640, 500, 0, 0, 0]

plt.plot(years, kohli, color='black', linestyle='--', marker='^', label="Kohli")
plt.plot(years, rohit, color='orange', linestyle=':', marker='x', label="Rohit")
plt.plot(years, sehwag, color='blue', linestyle='-', marker='o', label="Sehwag")
plt.legend()
plt.style.use("ggplot")
```
