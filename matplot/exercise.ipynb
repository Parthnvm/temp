```python
import matplotlib.pyplot as plt

products = ["Product Alpha", "Product Bravo", "Product Charlie", "Product Delta"]
sales_k = [120, 150, 95, 135] # Sales in Thousands of $

plt.barh(products,sales_k,color='blue')

for i, value in enumerate(sales_k):
    plt.text(value+ 2, i, f"${value}K", va='center')
plt.xlabel("Sales (in $K)")
plt.title("Quarterly Product Sales")

```

```python
import matplotlib.pyplot as plt
import numpy as np

years = ["Product Alpha", "Product Bravo", "Product Charlie", "Product Delta"]
q1_sales = [110, 140, 100, 150]
q2_sales = [120, 150, 95, 135] # The data from the first exercise

x = np.arange(len(years))
width = 0.35

plt.bar(x - width/2, q1_sales, width=width, label='Q1 Sales' ,color='blue')
plt.bar(x + width/2, q2_sales, width=width, label='Q2 Sales' ,color='black')

plt.xticks(x,years)
plt.xlabel("Sales (in $K)")
plt.title("Quarterly Product Sales Comparison")

plt.tight_layout()
```
